```python
# Membuat objek Radius Neighbors Classifier
radius = 0.7  # Nilai radius yang diinginkan
clf = RadiusNeighborsClassifier(radius=radius)

#Split Dataset 
X = data.drop('price_range',axis = 1)
y = data['price_range']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)

# Menyimpan model RadiusNeighbors ke file menggunakan joblib
joblib.dump(clf, 'radiusneighbors_model.pkl')

# Menampilkan pesan setelah menyimpan model
print("Model Radius Neighbors telah disimpan")
```
