**Mengimport Library yang dibutuhkan**

```python
# Mengimpor library yang diperlukan
from sklearn.neighbors import RadiusNeighborsClassifier
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

**Memuat Data**

```python
# Memuat dataset 
df = pd.read_csv('train.csv')
# Menampilkan Dimensi dataset
df.shape
```

**Memilih Fitur**

```python
# membuang kolom 'price_range' dari dataframe
X = df.drop(['price_range'], axis=1) 
# hanya mengambil kolom 'price_range' dari dataframe
y = df['price_range']
```

**Mencari Data Yang Hilang**

```python
# Mendeteksi missing value dalam DataFrame
missing_values = df.isnull().sum()

print(missing_values)
```

**Split dataset menjadi dua yaitu menjadi data train dan data test**

```python
# Membuat data training dan data testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# cek bentuk dari X_train and X_test
X_train.shape, X_test.shape
```

**Training Menggunakan Radius Neighbor**

```python
clf = RadiusNeighborsClassifier(radius=450)

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
y_pred

# mengecek akurasi prediksi
acc_dt = accuracy_score(y_pred, y_test)*100
print('Akurasi Model : ' + str(round(acc_dt, 2)) + ' %.')

```

```python
# report keseluruhan
report = classification_report(y_pred, y_test)
print(report)
```

```python
# save Model
joblib.dump(clf, 'radiusneighbors_model.pkl')
```
